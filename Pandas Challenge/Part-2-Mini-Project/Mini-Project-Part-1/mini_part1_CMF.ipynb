```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csvpath= "Resources/Coders_Survey_Data.csv"
```

```python
# Read with pandas
data = pd.read_csv(csvpath)
data.head()
```

```python
# Inspect all columns
data.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)
data = data.iloc[:, [0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]

data.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
# Hint use the df.replace function
data["AttendedBootcamp"] = data["AttendedBootcamp"].replace({0:'No', 1:'Yes'})
data.head()
```

```python
# Calculate total number of respondents in survey
# consider using the len() function
total_respondents= len(data)
```

```python
# Extract rows corresponding only to people who attended a bootcamp
# Filter using df.loc on the AttendedBootcamp column
data_attendees= data.loc[data["AttendedBootcamp"]== 'Yes']
data_attendees.head()
```

```python
# Calculate average age of attendees
# Pull out the `Age` series and take the average of it
avg_age= data_attendees["Age"].mean()
avg_age
```

```python
# Calculate how many people attended a bootcamp
# Count the values of the AttendedBootcamp column
attended_bc=data_attendees['AttendedBootcamp'].count()
attended_bc
```

```python
# Calculate how many attendees hold degrees
# It is easier to do this with `value_counts() count values for the SchoolDegree column
data_attendees["SchoolDegree"].value_counts()
```

```python
# Count the number of records where the person is a degree holder
# There are several ways to approach this. You can look for people who have degrees
# or for people who don't have degrees depending on the value of the SchoolDegree column
data_attendees ["SchoolDegree"].count()
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
gender_number =data_attendees['Gender'].value_counts()
gender_number
```

```python
# Calculate percentage of respondents who attended a bootcamp
bootcamp_avg = (len(data_attendees)/ len(data))
bootcamp_avg
```

```python
# Calculate percentage of respondents belonging to each gender
gender_pct= data_attendees["Gender"].value_counts("male")
male_pct= gender_pct[0]
female_pct= gender_pct[1]
non_specific_pct = gender_pct[2]+gender_pct[3]+gender_pct[4]
non_specific_pct
```

```python
# Calculate percentage of attendees with a college degree
hold_degree_df=data_attendees["SchoolDegree"].value_counts()
hold_degree_df

total_degree=hold_degree_df.loc[["bachelor's degree","master's degree (non-professional)","professional degree (MBA, MD, JD, etc.)","associate's degree","Ph.D." ]].sum()

degree_pct= total_degree / attended_bc
```

```python
# Calculate average post-bootcamp salary

postsalary_df = data["BootcampPostSalary"].mean()
postsalary_df
```

```python
# Create a new table consolidating above calculations

```

```python
new_bootcamp=pd.DataFrame ({"Total Surveyed": [total_respondents], "Total Bootcamp Attendees": [attended_bc],
                           " % Attended Bootcamp":[bootcamp_avg], "Avg Age": [avg_age], "Has a Degree": degree_pct, 
                            "% Male": male_pct, "% Female": female_pct, "% Non Gender Specific": non_specific_pct, 
                            "Average Post Bootcamp Salary": postsalary_df })
new_bootcamp
```

```python
# Improve formatting before outputting spreadsheet

new_bootcamp["% Male"]= new_bootcamp["% Male"].map("{:.2%}".format)
new_bootcamp["% Female"]= new_bootcamp["% Female"].map("{:.2%}".format)
new_bootcamp["% Non Gender Specific"]= new_bootcamp["% Non Gender Specific"].map("{:.2%}".format)
new_bootcamp["Has a Degree"]= new_bootcamp["Has a Degree"].map("{:.2%}".format)
new_bootcamp["Average Post Bootcamp Salary"]= new_bootcamp["Average Post Bootcamp Salary"].map("${:,.0f}".format)
new_bootcamp["Avg Age"]= new_bootcamp["Avg Age"].map("{:,.2f}".format)
new_bootcamp[" % Attended Bootcamp"]= new_bootcamp[" % Attended Bootcamp"].map("{:.2%}".format)
new_bootcamp
```

```python
# Export to Excel
# Use df.to_excel to export to excel. Don't include the indexes
new_bootcamp.to_excel ("output/BootcampOutputTask1.xlsx", index = False)
```
